**spaCy - Análise sintática**

```python
# importando a biblioteca
import spacy

# instanciando o objeto 
nlp = spacy.load('pt_core_news_lg')

# instanciando o texto
corpus = open('/content/corpus_teste.txt').read()

# treinando o modelo no texto
doc = nlp(corpus)

# realizando análise sintática
analise_sintatica = [(token.orth_, token.dep_) for token in doc]
analise_sintatica[:15]
```

```python
# visualizando a árvore sintática

# importando as libs e func
import spacy
from pathlib import Path  

# instanciando o objeto
nlp = spacy.load('pt_core_news_lg')

# instanciando o texto
corpus = open('/content/corpus_teste.txt').read()

# treinando o objeto no texto
doc = nlp(corpus)

# instanciando a árvore
svg = spacy.displacy.render(doc, style='dep')
output_path = Path('analise_dependencias.svg')
output_path.open('w', encoding='utf-8').write(svg)
```
