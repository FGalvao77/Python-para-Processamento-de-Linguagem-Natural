```python
# criando uma função para otimizar o tratamento do arquivo de texto

def tratamento_texto(corpus, top_freq=None):

    # importando a função "regex" para tokenização
    from nltk.tokenize import RegexpTokenizer

    # instanciando a função "regex" na variável "tokenizer"
    tokenizer = RegexpTokenizer(r'[a-zA-Z]\w*') # ignorando todos os símbolos e números

    # aplicando a tokenização e salvando o resultado na variável "tokens"
    tokens = tokenizer.tokenize(corpus)

    # instanciando a lista de stopwords na variável "stopwords"
    stopwords = nltk.corpus.stopwords.words('portuguese')

    # criando uma list comprehesion para tratar o arquivo de texto
    texto_tratado = [token.lower() for token in tokens if token.lower() not in stopwords]

    # imprimindo as frequências das palavras
    freq = nltk.FreqDist(texto_tratado)
    top_palavras = freq.most_common(top_freq)  # visualizando as palavras mais frequentes

    return top_palavras
```

```python
# contabilizando os tokens
tratamento_texto(corpus=open('/content/corpus_teste.txt').read())
```

```python
# contabilizando os tokens - os 15 mais frequentes
tratamento_texto(corpus=open('/content/corpus_teste.txt').read())[:15]
```

```python
# criando uma função para otimizar o tratamento do arquivo de texto

def word_processing(corpus, word_frequencies=None):

    # importando a função "regex" para tokenização
    from nltk.tokenize import RegexpTokenizer

    # instanciando a função "regex" na variável "tokenizer"
    tokenizer = RegexpTokenizer(r'[a-zA-Z]\w*') # ignorando todos os símbolos e números

    # aplicando a tokenização e salvando o resultado na variável "tokens"
    tokens = tokenizer.tokenize(corpus)

    # instanciando a lista de stopwords na variável "stopwords"
    stopwords = nltk.corpus.stopwords.words('portuguese')

    # criando uma list comprehesion para tratar o arquivo de texto
    treated_text = [token.lower() for token in tokens if token.lower() not in stopwords]

    # imprimindo as frequências das palavras
    frequencies = nltk.FreqDist(treated_text)
    top_of_words = frequencies.most_common(word_frequencies)  # visualizando as palavras mais frequentes

    return top_of_words
```

```python
# criando uma função para otimizar o tratamento do arquivo de texto

def word_processing(corpus, language=stopwords, word_frequencies=None):
    '''
    Essa função tem a finalidade em tratar textos em geral gerando a tokenização do mesmo.
    Excluindo símbolos de acentuação e caracteres especiais, números e deixando as 
    palavras em caixa baixa.

    Como argumentos da função, devemos passar:
    - corpus (obrigatório): o arquivo de texto;
    - language (obrigatório): a língua que será realizado o stopwords e,
    - word_frequencies (não obrigatório): você pode definir o ranque (quais mais frequentes palavras exibirá).
        Se não passar nemhum número do ranque, será exibido todas as frequências.
    '''

    # importando a função "regex" para tokenização
    from nltk.tokenize import RegexpTokenizer

    # instanciando a função "regex" na variável "tokenizer"
    tokenizer = RegexpTokenizer(r'[a-zA-Z]\w*') # ignorando todos os símbolos, números e deixando as palavras em caixa baixa

    # aplicando a tokenização e salvando o resultado na variável "tokens"
    tokens = tokenizer.tokenize(corpus.read())

    # instanciando a lista de stopwords na variável "stopwords"
    stopwords = nltk.corpus.stopwords.words(language)

    # criando uma list comprehesion para tratar o arquivo de texto
    treated_text = [token.lower() for token in tokens if token.lower() not in stopwords]

    # imprimindo as frequências das palavras
    frequencies = nltk.FreqDist(treated_text)
    top_of_words = frequencies.most_common(word_frequencies)  # visualizando as palavras mais frequentes

    return top_of_words
```

```python
from nltk.corpus import mac_morpho
from nltk.tag import UnigramTagger
from nltk.tag import DefaultTagger
```

```python
tokens = nltk.word_tokenize(corpus)
```

```python
etiquetador_padrao = DefaultTagger('N')

sentencas_treino = mac_morpho.tagged_sents()
etiquetador = UnigramTagger(sentencas_treino, backoff=etiquetador_padrao)
```

```python
etiquetado = etiquetador.tag(tokens)
etiquetado
```

```python
etiquetado[:15]
```
